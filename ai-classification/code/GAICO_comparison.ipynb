```python
import numpy
numpy.version.version
```

```python
from gaico import Experiment
```

```python
# Test Case 1 Short Positive 
llm_responses = {
    "GPT": "positive",
    "Fine-tuned Distilbert": "positive",
    "Base Distilbert": "negative",
    "Logistic Regression": "positive",
}
reference_answer = "positive"
```

```python
exp = Experiment(
    llm_responses=llm_responses,
    reference_answer=reference_answer
)
```

```python
results_df = exp.compare(
    metrics=['Jaccard', 'ROUGE'], 
    plot=True,                     
    output_csv_path="testcase1_report.csv",
    custom_thresholds={"Jaccard": 0.6, "ROUGE_rouge1": 0.35}
)
```
